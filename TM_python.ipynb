````python
# save html files in folder "html_google_scholar", each file has 20 papers
# use BeautifulSoup to read the content of files and find the title and url(link) of paper
results = []
for file in os.listdir("html_google_scholar"):
    f=open('html_google_scholar/{}'.format(file), encoding="utf8")
    page = BeautifulSoup(f.read(), 'lxml')
    for entry in page.find_all("h3", attrs={"class": "gs_rt"}):
        results.append({"title": entry.a.text, "url": entry.a['href']})
````

### text cleaning to all papers

````python
def preprocess_text(text):
    # remove extra spaces
    text = re.sub("\s+"," ",text)
    
    # remove punctuations
    text = re.sub("[^0-9A-Za-z ]", "" , text)
    
    #case normalization
    text = "".join([i.lower() for i in text])

    #tokenizaiton
    tokens = nltk.tokenize.word_tokenize(text)
    
    #remove stopwords
    stopwords = nltk.corpus.stopwords.words('english')
    tokens = [token for token in tokens if token not in stopwords]
    
    #lemmatization
    wn = nltk.WordNetLemmatizer()
    tokens = [wn.lemmatize(token) for token in tokens]
    
    return " ".join(tokens)
````

````python
files = os.listdir("html_google_scholar/papers")
files = [int(os.path.splitext(filename)[0]) for filename in files]
files.sort()
for Nr in tqdm(files):
    file_path = "html_google_scholar/papers_text/{}.txt".format(Nr)
    with open(file_path, encoding='utf-8') as f:
        text = f.read()
        
    preprocessed_text = preprocess_text(text)
    
    output_file_path = "html_google_scholar/preprocessed_text/{}.txt".format(Nr)
    with open(output_file_path, "w",encoding="utf-8") as f:
        f.write(preprocessed_text)
````

# Text Transformation (Features Generation)

## Bag-of-Words (BoW) representation

each row represents a document and each column represents a word in the vocabulary

````python
# convert the 
files = os.listdir("html_google_scholar/preprocessed_text")
files = [int(os.path.splitext(filename)[0]) for filename in files]
files.sort()
docs = []
for Nr in tqdm(files):
    tokens_path = "html_google_scholar/preprocessed_text/{}.txt".format(Nr)
    with open(tokens_path, encoding='utf-8') as f:
        tokens = f.read()
    docs.append(tokens)
````

````python
# Create a CountVectorizer object
bow_vectorizer = CountVectorizer()

# Fit the vectorizer to the documents
bow = bow_vectorizer.fit_transform(docs)

# Get the vocabulary
vocab = bow_vectorizer.vocabulary_

# Convert the sparse matrix to a dense matrix
bow = bow.toarray()

# Create a dataframe from the bag-of-words representation
bow_df = pd.DataFrame(bow, columns=vocab)
````

##  term frequency-inverse document frequency (TF-IDF)

Evaluate the importance of a term in a document compared to a corpus (collection of documents).    
The idea behind TF-IDF is that words that are more frequent in a document but less frequent across the corpus are more informative and therefore more important to the document's meaning.

````python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(docs)
````

````python
# Get the feature names (vocabulary)
feature_names = tfidf_vectorizer.get_feature_names()

# Convert the sparse matrix to a dense matrix
dense = tfidf_matrix.todense()

# Convert the dense matrix to a DataFrame
tfidf_df = pd.DataFrame(dense, columns=feature_names)

# Get the vocabulary
vocab = tfidf_vectorizer.vocabulary_

# Convert the dense matrix to a DataFrame
tfidf_df = pd.DataFrame(dense, columns=vocab)

tfidf_df
````

````python
# initialize the TfidfVectorizer
vectorizer = TfidfVectorizer()

# fit and transform the documents to get the TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(docs)

# get the feature names (terms) from the vectorizer
terms = vectorizer.get_feature_names()

# create a pandas DataFrame to store the TF-IDF matrix
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=terms)

# iterate over the rows of the TF-IDF matrix and get the keywords for each document
for index, row in tfidf_df.iterrows():
    print(f"Document {index+1}:")
    keywords = [i for i in row.nlargest(5).index]
    print(f"Keywords: {keywords}")
````
