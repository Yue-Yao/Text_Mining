### text cleaning to all papers

````python
def preprocess_text(text):
    # remove extra spaces
    text = re.sub("\s+"," ",text)
    
    # remove punctuations
    text = re.sub("[^0-9A-Za-z ]", "" , text)
    
    #case normalization
    text = "".join([i.lower() for i in text])

    #tokenizaiton
    tokens = nltk.tokenize.word_tokenize(text)
    
    #remove stopwords
    stopwords = nltk.corpus.stopwords.words('english')
    tokens = [token for token in tokens if token not in stopwords]
    
    #lemmatization
    wn = nltk.WordNetLemmatizer()
    tokens = [wn.lemmatize(token) for token in tokens]
    
    return " ".join(tokens)
````

````python
files = os.listdir("html_google_scholar/papers")
files = [int(os.path.splitext(filename)[0]) for filename in files]
files.sort()
for Nr in tqdm(files):
    file_path = "html_google_scholar/papers_text/{}.txt".format(Nr)
    with open(file_path, encoding='utf-8') as f:
        text = f.read()
        
    preprocessed_text = preprocess_text(text)
    
    output_file_path = "html_google_scholar/preprocessed_text/{}.txt".format(Nr)
    with open(output_file_path, "w",encoding="utf-8") as f:
        f.write(preprocessed_text)
````

## Bag-of-Words (BoW) representation

each row represents a document and each column represents a word in the vocabulary

````python
# convert the text to BoW Vector
files = os.listdir("html_google_scholar/preprocessed_text")
files = [int(os.path.splitext(filename)[0]) for filename in files]
files.sort()
docs = []
for Nr in tqdm(files):
    tokens_path = "html_google_scholar/preprocessed_text/{}.txt".format(Nr)
    with open(tokens_path, encoding='utf-8') as f:
        tokens = f.read()
    docs.append(tokens)
````

````python
# Create a CountVectorizer object
bow_vectorizer = CountVectorizer()

# Fit the vectorizer to the documents
bow = bow_vectorizer.fit_transform(docs)

# Get the vocabulary
vocab = bow_vectorizer.vocabulary_

# Convert the sparse matrix to a dense matrix
bow = bow.toarray()

# Create a dataframe from the bag-of-words representation
bow_df = pd.DataFrame(bow, columns=vocab)
````

##  term frequency-inverse document frequency (TF-IDF)

Evaluate the importance of a term in a document compared to a corpus (collection of documents).    
The idea behind TF-IDF is that words that are more frequent in a document but less frequent across the corpus are more informative and therefore more important to the document's meaning.

````python
# initialize the TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# fit and transform the documents to get the TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(docs)
````

````python
# Get the feature names (vocabulary)
feature_names = tfidf_vectorizer.get_feature_names()

# Convert the sparse matrix to a dense matrix
dense = tfidf_matrix.todense()

# Convert the dense matrix to a DataFrame
tfidf_df = pd.DataFrame(dense, columns=feature_names)

# Get the vocabulary
vocab = tfidf_vectorizer.vocabulary_

# Convert the dense matrix to a DataFrame
tfidf_df = pd.DataFrame(dense, columns=vocab)

tfidf_df
````

````python
# create a pandas DataFrame to store the TF-IDF matrix
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

# iterate over the rows of the TF-IDF matrix and get the keywords for each document
for index, row in tfidf_df.iterrows():
    print(f"Document {index+1}:")
    keywords = [i for i in row.nlargest(5).index]
    print(f"Keywords: {keywords}")
````

### Doc2Vec
use a vector of numbers to represent each document in corpus, the number of dimension is manually defined

````python
# convert the text to Vector
files = os.listdir("html_google_scholar/preprocessed_text")
files = [int(os.path.splitext(filename)[0]) for filename in files]
files.sort()
docs = []
for Nr in tqdm(files):
    tokens_path = "html_google_scholar/preprocessed_text/{}.txt".format(Nr)
    with open(tokens_path, encoding='utf-8') as f:
        tokens = f.read()
    docs.append(tokens)
````

````python
#data to be trained
data = docs
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(data)]

# traning setting
max_epochs = 40
vec_size = 20
alpha = 0.025

model = Doc2Vec(vector_size=vec_size,
                alpha=alpha,
                min_alpha=0.00025,
                min_count=1,
                dm=1)

'''
Note: dm defines the training algorithm. If dm=1 means‘distributed memory’(PV-DM) and dm =0 means‘distributed bag of words’(PV-DBOW). 
Distributed Memory model preserves the word order in a document whereas Distributed Bag of words just uses the bag of words approach, 
which doesn’t preserve any word order.
'''

model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.epochs)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha

model.save("d2v.model")
print("Model Saved")
````

# Feature Selection
Use clustering(Kmeans) to group the documents into different clusters based on their feature similarity. Then print 10 features in each group with highest mean TF-IDF score

````python
# Convert corpus to document-term matrix using TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english',min_df=5)
X = vectorizer.fit_transform(docs)

# Use KMeans clustering to group documents into 5 clusters
kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
labels = kmeans.labels_

# Extract feature names from vectorizer
feature_names = vectorizer.get_feature_names()

# Create list of arrays where each array contains the indices of the data points that belong to the corresponding cluster
clusters = []
for i in range(0, 5):
    cluster_indices = np.where(labels == i)[0]
    clusters.append(cluster_indices)

# Calculate the mean TF-IDF score for each feature in each cluster
mean_tfidf_scores = []
for i in range(0, 5):
    cluster_mean_tfidf_scores = X[clusters[i]].mean(axis=0)
    cluster_mean_tfidf_scores = np.asarray(cluster_mean_tfidf_scores).reshape(-1)
    sorted_scores = sorted(zip(feature_names, cluster_mean_tfidf_scores), key=lambda x: x[1], reverse=True)
    mean_tfidf_scores.append(sorted_scores)

# Print the top 10 features with the highest mean TF-IDF score for each cluster
for i in range(0, 5):
    print('Cluster {}:'.format(i))
    for feature, score in mean_tfidf_scores[i][:10]:
        print('\t{}: {:.3f}'.format(feature, score))
````

# Topic Modeling
identify the main topics that are discussed using Latent Dirichlet Allocation (LDA) model

````python
# convert the text to Vector
files = os.listdir("html_google_scholar/preprocessed_text")
files = [int(os.path.splitext(filename)[0]) for filename in files]
files.sort()
docs = []
for Nr in tqdm(files):
    tokens_path = "html_google_scholar/preprocessed_text/{}.txt".format(Nr)
    with open(tokens_path, encoding='utf-8') as f:
        tokens = f.read()
    docs.append(tokens)
````

````python
long_string = ' '.join(docs)
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=10000, contour_width=3, contour_color='steelblue')
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
wordcloud.to_image()
````

````python
# Load the corpus of documents
doc_corpus = docs

# Tokenize the documents and create a dictionary
doc_tokens = [[token for token in doc.split()] for doc in doc_corpus]
dictionary = corpora.Dictionary(doc_tokens)

# Convert the tokenized documents into bag-of-words format
doc_bow = [dictionary.doc2bow(doc) for doc in doc_tokens]

# Train the LDA model on the bag-of-words representation of the documents
lda_model = LdaModel(doc_bow, num_topics=10, id2word=dictionary, passes=20)

# Print the topics and the top 5 words for each topic
for topic in lda_model.print_topics(num_topics=10, num_words=5):
    print(topic)
````

````python

# Create a visualization of the LDA model
vis = pyLDAvis.gensim_models.prepare(lda_model, doc_bow, dictionary)

# Display the visualization in a web browser
pyLDAvis.display(vis)
````

- Each bubble represents a topic. The larger the bubble, the higher percentage of the number of documents in the corpus is about that topic.  
- Blue bars represent the overall frequency of each word in the corpus. If no topic is selected, the blue bars of the most frequently used words will be displayed.
- Red bars give the estimated number of times a given term was generated by a given topic. There are about 56,000 of the word ‘li’, and this term is used about 3,000 times within topic 1. The word with the longest red bar is the word that is used the most by the documents belonging to that topic.
- λ controls the balance between two types of topic relevance measures
    - λ=1, emphasizes the importance of the most **frequent** words in the topic, and may be useful for exploratory analysis of the topics.
    - λ=0, emphasizes the importance of the most **specific** words in the topic, and may be useful for **more focused analysis** of the topics.

Numbers represent the weights of each word in the topics learned by the LDA model, also the most important word in texts
