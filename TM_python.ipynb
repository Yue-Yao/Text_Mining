# Collect Papers

## .html files from webs

````python
# save html files in folder "html_google_scholar", each file has 20 papers
# use BeautifulSoup to read the content of files and find the title and url(link) of paper
results = []
for file in os.listdir("html_google_scholar"):
    f=open('html_google_scholar/{}'.format(file), encoding="utf8")
    page = BeautifulSoup(f.read(), 'lxml')
    for entry in page.find_all("h3", attrs={"class": "gs_rt"}):
        results.append({"title": entry.a.text, "url": entry.a['href']})
````

````python
results
````

## download pdf files using url links

````python
# use API from SciHub to get pdf files named by numbers
# some papers cannot be downloaded because of captcha and request exception
# these papers should be manually acquired
sh = SciHub()
numbers = range(1,61)
for paper,num in zip(results,numbers):
    try: 
        dl = sh.download(paper['url'], path= 'html_google_scholar/papers/{}.pdf'.format(num))
        print('paper Nr.{} completed'.format(num))
    except:
        print('paper Nr.{} failed'.format(num))
````

````python
# sort by numbers
files = os.listdir("html_google_scholar/papers")
files = [int(os.path.splitext(filename)[0]) for filename in files]
files.sort()
````

````python
# get the list of papers and their download status
# save all this info in a excel file
paper_df = pd.DataFrame(results)
#set the index as the number of paper
paper_df.index+=1
#check if the paper downloaded
paper_df['download']=False
for i in paper_df.index:
    if i in files:
        paper_df.loc[[i],['download']]=True
paper_df.to_excel('html_google_scholar/paperlist.xlsx')
````

## extract info from pdf files

### text info

````python
# use PyMuPDF(fitz) to extract text from pdf
pdf = "html_google_scholar/papers/1.pdf"
doc = fitz.open(pdf)
for page in doc:
    text = page.get_text("text")
    print(text)
````

````python
# write text into .txt files, also do it using R code
files = os.listdir("html_google_scholar/papers")
files = [int(os.path.splitext(filename)[0]) for filename in files]
files.sort()
for Nr in files:
    pdf = "html_google_scholar/papers/{}.pdf".format(Nr)
    doc = fitz.open(pdf)
    with open("html_google_scholar/papers_text/{}.txt".format(Nr), "w",encoding="utf-8") as text_file:
        for page in doc:
            text = page.get_text("text")
            print("{}".format(text), file=text_file)
````

## text cleaning

### remove extra spaces 

````python
with open("html_google_scholar/papers_text/1.txt",encoding='utf-8') as text_file:
    text = text_file.read()
````

````python
print(text)
````

````python
text = re.sub("\s+"," ",text)
print(text)
````

### remove punctuations

````python
text = re.sub("[^-9A-Za-z ]", "" , text)# just keep the elements with at least one number 0-9 or letter A-Z or a-z
print(text)
````

### case normalization

````python
text = "".join([i.lower() for i in text])
print(text)
````

### tokenization

````python
nltk.download('punkt')
````

````python
nltk.tokenize.word_tokenize(text)
````

````python
len(nltk.tokenize.word_tokenize(text))
````

### remove stopwords

````python
nltk.download('stopwords')
````

````python
stopwords = nltk.corpus.stopwords.words('english')
text = [i for i in text.split() if i not in stopwords]
print(text)
````

````python
len(text)
````

````python
len(np.unique(text))
````

### lemmatization 

````python
nltk.download('wordnet')
````

````python
wn = nltk.WordNetLemmatizer()
word = [wn.lemmatize(word) for word in text]
print(word)
````

````python
len(np.unique(word))
````

````python
word_count = Counter(word)
word_count.most_common(10)
````
